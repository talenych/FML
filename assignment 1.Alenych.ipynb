```python
import seaborn as sns #this is the plotting library I'll be using 
import pandas as pd #"as pd" means that we can use the abbreviation in commands
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs

df1 = pd.read_csv('steps.csv',sep=';') #firsly we read the csv file and reduce ";" from there
df1.head()

```

```python
df2 = pd.read_csv('survey.csv') #firsly we read the csv file
df2.head()
```

```python
#A plot of a time series of my choice
df1['23-6-2013'].name = 'Steps' #Labeling the variable for plotting
file_steps = df1['23-6-2013'].dropna() #Selecting the size column. I need to drop the NA's, otherwise Seaborn won't plot
sns.distplot(file_steps, kde=False) 
plt.title('Mean steps per day over all participants on 23-6-2013') #creating a title for our plot
plt.ylabel('Participants')
plt.show()
```

```python
#Summary statistics
mean_steps = df1['23-6-2013'].mean()
print(f'The mean steps on 23-6-2013 is: {mean_steps} steps')

median_steps = df1['23-6-2013'].median()
print(f'The median steps is: {median_steps} steps')
```

```python
#1st graph
sns.violinplot(file_steps)
sns.swarmplot(file_steps, color='yellow') #default color doesn't stand out
plt.title('Mean steps per day over all participant on 23.06.2013') #creating a title for our plot
plt.ylabel('Participants')
plt.xlabel('Steps')
plt.show()
```

```python
#2nd graph
sns.distplot(file_steps, rug=True)
plt.title('Mean steps per day over all participant on 23.06.2013')
plt.xlabel('Steps')
plt.ylabel('Proportions')
plt.show()
```
